## Integrating Fawkes  with Google Colab

Here are the steps to get Fawkes running on Colab. Follow the steps along to get Fawkes running on Google's GPU's and leverage this resource for free! 

Login into your Google account and run the command below to make sure you are able to mount your drive from Colab. This will help you clone the project and build it as well. 

**How to load your private notebook on Colab:**

Once you have a forked copy of Fawkes, to load the notebook from your private GitHub repository do the following:

1. Navigate to http://colab.research.google.com/github.
2. Click the "Include Private Repos" checkbox.
3. In the popup window, sign-in to your Github account and authorize Colab to read the private files.
4. Your private repositories and notebooks will now be available via the github navigation pane.

You need to now first mount your Drive to this runtime. You see the next code block? Hit the play button to run it. Open the URL and proive your authorization code. On sucess you would see that it has been mounted.

In the next code block edit the `GIT_USERNAME` variable to that of your user named and the add in your `GIT_TOKEN` as well. Refer [this](https://docs.github.com/en/free-pro-team@latest/github/authenticating-to-github/creating-a-personal-access-token) to create a personal access token.

Check the ouptut of the print statements to verify or modify anything as needed.
The `PROJECT_PATH` will be the folder where in the repository will be cloned.

```
from os.path import join  

ROOT = '/content/gdrive' # default loaction on Drive
DRIVE_PATH = 'My Drive/projects' # path to your project on Google Drive, you can edit this to suit your needs!

PROJECT_PATH = join(ROOT, DRIVE_PATH)

GIT_USERNAME = "replace_with_your_git_username" 
GIT_TOKEN = "replace_with_your_personal_access_token"  

!mkdir "{PROJECT_PATH}" # a folder will be creared in the project path if it isnt there already!

GIT_PATH = "https://" + GIT_TOKEN + "@github.com/" + GIT_USERNAME + "/" + "fawkes.git"

print("PROJECT_PATH: ", PROJECT_PATH) # you can check the output of this if you are happy with the path you provided
print("GIT_PATH: ", GIT_PATH)
```

**How do I save notebooks to Drive or Github?**

Any time you open a GitHub hosted notebook in Colab, it opens a new editable view of the notebook. You can run and modify the notebook without worrying about overwriting the source.

If you would like to save your changes from within Colab, you can use the File menu to save the modified notebook either to Google Drive or back to GitHub. Choose **File→Save a copy in Drive** or **File→Save a copy to GitHub** and follow the resulting prompts. To save a Colab notebook to GitHub requires giving Colab permission to push the commit to your repository.
